```python
# Python

# ì–´ë– í•œ ë³€í™”ë„ ê°€í•˜ì§€ ì•Šì€ ížˆìŠ¤í† ê·¸ëž¨
plt.hist(x = EuStockMarkets['SMI'], bins = 30)
plt.grid()
plt.show()

# ì°¨ë¶„ì— ëŒ€í•œ ížˆìŠ¤í† ê·¸ëž¨
# ì¶”ì„¸ë¥¼ ì œê±°í•˜ê¸° ìœ„í•´ ì‹œê°„ìƒ ì¸ì ‘í•œ ë°ì´í„°ì˜ ì°¨ì´ë¥¼ êµ¬í•˜ë©´, ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” í˜•íƒœë¡œ ë³€í™”í•¨.
plt.hist(x = EuStockMarkets['SMI'].diff(), bins = 30)
plt.grid()
plt.show()
```

```python
#Python

# ì‹œê°„ì— ë”°ë¥¸ ì„œë¡œ ë‹¤ë¥¸ ë‘ ì£¼ì‹ì˜ ê°€ì¹˜
plt.scatter(x = EuStockMarkets['SMI'], y = EuStockMarkets['DAX'])
plt.grid()
plt.show()

# ì°¨ë¶„ì— ëŒ€í•œ ë‘ ì£¼ì‹ì˜ ì¼ì¼ ê°€ì¹˜ ë³€ë™
plt.scatter(x = EuStockMarkets['SMI'].diff(), y = EuStockMarkets['DAX'].diff())
plt.grid()
plt.show()
```

```python
# Python

# shift(=lag) í•¨ìˆ˜ ì ìš©í•œ ì‚°ì ë„
# ë‘ ì£¼ê°€ì˜ ìƒê´€ê´€ê³„ê°€ ì‚¬ë¼ì§
# ì¦‰ í•œ ë³€ìˆ˜ë¡œ ë‹¤ë¥¸ ë³€ìˆ˜ë¥¼ ì˜ˆì¸¡í•˜ì§€ ëª»í•¨
plt.scatter(x = EuStockMarkets['SMI'].diff().shift(1),
            y = EuStockMarkets['DAX'].diff())
plt.grid()
plt.show()
```
